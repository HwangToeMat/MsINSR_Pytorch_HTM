```python
from MsINSR import MsINSR_Net, D_Net
import torch.backends.cudnn as cudnn
from dataset_h5 import Read_dataset_h5
from torch.utils.data import DataLoader
from PIL import Image
import torch
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
G1 = MsINSR_Net().eval().float()
checkpoint = torch.load('checkpoint/MsINSR_Adam_epoch_1.tar')
G1.load_state_dict(checkpoint['G_Net_state_dict'])
G_10 = G1.cpu()

G2 = MsINSR_Net().eval().float()
checkpoint = torch.load('checkpoint/MsINSR_Adam_epoch_3.tar')
G2.load_state_dict(checkpoint['G_Net_state_dict'])
G_15 = G2.cpu()

G3 = MsINSR_Net().eval().float()
checkpoint = torch.load('checkpoint/MsINSR_Adam_epoch_5.tar')
G3.load_state_dict(checkpoint['G_Net_state_dict'])
G_20 = G3.cpu()
```

```python
test_img = cv2.imread('/content/gdrive/My Drive/SRGAN_Pytorch_HTM/data/Test/Set5/khu_4.jpg')
input_img = torch.tensor(test_img.reshape(1,3,test_img.shape[0],test_img.shape[1])/255)
```

```python
output1 = G_10(input_img.float())
output2 = G_15(input_img.float())
output3 = G_20(input_img.float())
```

```python
output_img = output1*255
output_img = np.round(output_img.detach().numpy()).astype(int)
output_img1 = output_img.reshape(output_img.shape[2],output_img.shape[3],output_img.shape[1])
cv2.imwrite('result/test_data_epoch_5.png', output_img1)

output_img = output2*255
output_img = np.round(output_img.detach().numpy()).astype(int)
output_img2 = output_img.reshape(output_img.shape[2],output_img.shape[3],output_img.shape[1])
cv2.imwrite('result/test_data_epoch_10.png', output_img2)

output_img = output3*255
output_img = np.round(output_img.detach().numpy()).astype(int)
output_img3 = output_img.reshape(output_img.shape[2],output_img.shape[3],output_img.shape[1])
cv2.imwrite('result/test_data_epoch_15.png', output_img3)
```

```python
fig = plt.figure(figsize=(15, 15))
ax = plt.subplot("221")
ax.imshow(test_img)
ax.set_title("GT_img")

ax = plt.subplot("222")
ax.imshow(output_img1)
ax.set_title("output_img(SRGAN)_epoch_10")

ax = plt.subplot("223")
ax.imshow(output_img2)
ax.set_title("output_img(SRGAN)_epoch_15")

ax = plt.subplot("224")
ax.imshow(output_img3)
ax.set_title("output_img(SRGAN)_epoch_20")
```
